```python
dst_path = "/home/krri/brtdata/2021_06_BAG/raw/Train_Test Dataset/Train/"
timepath_list = os.listdir("/home/krri/brtdata/2021_06_BAG/raw")
timepath_list.remove("Disable")
timepath_list.remove("Train_Test Dataset")

for timepath in timepath_list:
    ori_path = f"/home/krri/brtdata/2021_06_BAG/raw/{timepath}/sync_5_sec/"
    c1_list = os.listdir(ori_path + "c1")
    c2_list = os.listdir(ori_path + "c2")
    l1_list = os.listdir(ori_path + "l1")
    l2_list = os.listdir(ori_path + "l2")

    
    for c1,c2,l1,l2 in zip(c1_list,c2_list,l1_list,l2_list): 
        ori_c1 = ori_path + "c1/" + c1
        ori_c2 = ori_path + "c2/" + c2
        ori_l1 = ori_path + "l1/" + l1
        ori_l2 = ori_path + "l2/" + l2
        
        dst_c1 = dst_path + name_dict[timepath[8:]] + "/c1/21" + name_dict[timepath[8:]] + "C1_" + c1
        dst_c2 = dst_path + name_dict[timepath[8:]] + "/c2/21" + name_dict[timepath[8:]] + "C2_" + c2
        dst_l1 = dst_path + name_dict[timepath[8:]] + "/l1/21" + name_dict[timepath[8:]] + "L1_" + l1
        dst_l2 = dst_path + name_dict[timepath[8:]] + "/l2/21" + name_dict[timepath[8:]] + "L2_" + l2
        
        if not os.path.isdir(dst_path + name_dict[timepath[8:]] + "/c1"):
            os.makedirs(dst_path + name_dict[timepath[8:]] + "/c1")
        if not os.path.isdir(dst_path + name_dict[timepath[8:]] + "/c2"):
            os.makedirs(dst_path + name_dict[timepath[8:]] + "/c2")
        if not os.path.isdir(dst_path + name_dict[timepath[8:]] + "/l1"):
            os.makedirs(dst_path + name_dict[timepath[8:]] + "/l1")
        if not os.path.isdir(dst_path + name_dict[timepath[8:]] + "/l2"):
            os.makedirs(dst_path + name_dict[timepath[8:]] + "/l2")
            
        shutil.copyfile(ori_c1,dst_c1)
        shutil.copyfile(ori_c2,dst_c2)
        shutil.copyfile(ori_l1,dst_l1)
        shutil.copyfile(ori_l2,dst_l2)
        
        print(dst_c1,end="\r")
```
