```python
basePath = "/home/krri/brtdata/exp_20201111/raw/Train_Test Dataset/Scenario/"
usecase_list = df["유스케이스 번호"].tolist()
time_list = df["날짜 시간"].tolist()
start_frame_list = df["시작 프레임"].tolist()
end_frame_list = df["끝 프레임"].tolist()

for time, usecase, start_frame, end_frame in zip(time_list,usecase_list, start_frame_list, end_frame_list):
    print(usecase)
    dst_concatPath = os.path.join(basePath,usecase,"concat_img/")
    if not(os.path.exists(dst_concatPath)):
        os.makedirs(dst_concatPath)

    ori_concatPath = f"/home/krri/brtdata/exp_20201111/raw/{time}/sync/concat_img/"
    for i in range(start_frame, end_frame):
        file_name = str(i).zfill(6)+".jpg"
        print(f"Copying >> {ori_concatPath + file_name}",end="\r")
        shutil.copyfile(ori_concatPath + file_name, dst_concatPath + file_name)
```
