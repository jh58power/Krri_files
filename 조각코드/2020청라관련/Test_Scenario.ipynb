```python
usecase_list = df["유스케이스 번호"].tolist()
basePath_list = df["날짜 시간"].tolist()
start_frame_list = df["시작 프레임"].tolist()
end_frame_list = df["끝 프레임"].tolist()
```

```python
save_path = "/home/krri/brtdata/exp_20201111/raw/Train_Test Dataset/Scenario/"

for i in usecase_list:
    for j in range(1,5):
        if (not os.path.isdir(save_path + i + f"/c{j}/")):
            os.makedirs(save_path + i + f"/c{j}/")

        if (not os.path.isdir(save_path + i + "/ld/")):
            os.makedirs(save_path + i + "/ld/")
```
